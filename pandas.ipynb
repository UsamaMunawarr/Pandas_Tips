## 16-Splitting a string into multiple columns

```python
#import libraries
import pandas as pd
df = pd.DataFrame({'name':['Abu Usama','Ahmed Rasheed','Umar Twise','Hafiz Haroon','Zulquarnain Ali'],
                   'location':['Lhr_Pak','Grw_Pak','Fsd_Pak','IsL_Pak','Karaachi_Pak']})
df.head()
```

```python
df[["first_name","second_name"]] = df.name.str.split(' ',expand=True)
df
```

```python
df[["first_name","second_name"]] = df.name.str.split(' ',expand=True)
df[["City","Country"]] = df.location.str.split('_',expand=True)
df
```

```python
#Refine Data manipulation
df = df[['first_name','second_name','City','City']]
df
```
